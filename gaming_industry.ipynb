### 2. Предобработка данных

В первую очередь, приведем cтолбцы к нижнему регистру

```python
games_data.columns = games_data.columns.str.lower()
games_data.columns
```

`year_of_release` — методом describe мы увидели, что год выпуска находится в диапазоне 1980-2016. Заменить пропуски средним или медианным значением не представляется возможным. Замена приведет к искажению данных. Всего в столбце пропущено 1.61% строк. Можем удалить их, это не станет критичным для анализа.

```python
games_data = games_data.dropna(subset=['genre'])
```

```python
games_data['year_of_release'] = games_data['year_of_release'].astype('int')
```

Далее столбцы с оценками.

`critic_score` и `user_score` — субъективные показатели. Нельзя заменять пропуски средним или медианным значением, а также нулями (оценка "0" также важна). Оставим как есть. 

Сами пропуски могут быть обусловлены различными факторами. Возможно, не было самих опросов, из которых бы сформировались пользовательские оценки и оценки критиков. Пропуски могли появиться и по вине недостаточной популярности игры или из-за технического сбоя.

```python
games_data['user_score'].unique() 
```

Один нюанс: в столбце `user_score` присутствует аббревиатура `tbd`, что означает от английского To Be Decided (будет решено). Заменим значение это значение на NaN, чтобы при дальнейшей конвертации в float не случилась ошибка

```python
games_data['user_score'] = games_data['user_score'].replace('tbd', np.NaN)
games_data['user_score'] = games_data['user_score'].astype('float')
```

В работе со столбцом `rating` обратимся к расшифровке рейтинга ESRB (https://www.ph4.ru/pal_picto.php?p=ratings-esrb&titl=ESRB):

* «**EC**» («Early childhood») — «Для детей младшего возраста»: игра подходит для детей от 3 лет. Продукты, получившие данный рейтинг, изначально разрабатываются для детей и обычно представляют собой развивающие игры.
   
     
* «**E**» («Everyone») — «Для всех»: содержание вполне подходит для всех возрастов. Такие игры могут понравиться и взрослым. Игры с этим рейтингом могут содержать минимальное насилие, в основном «мультяшного» характера. 
   
     
* «**E10**» («Everyone 10 and older») — «Для всех от 10 лет и старше»: проекты с данным рейтингом могут содержать больше мультипликационного или мягкого насилия, или несколько откровенные сцены, или минимальное количество крови. Рейтинг был принят ESRB 2 марта 2005 года.
  
    
* «**T**» («Teen») — «Подросткам»: игра подходит для лиц от 13 лет. Проекты из данной категории могут содержать насилие, непристойные сцены, грубый юмор, кровь или нечастое использование ненормативной лексики.
  
    
* «**M**» («Mature») — «Для взрослых»: материалы игры не подходят для лиц младше 17 лет. Проекты с данным рейтингом могут содержать жестокое насилие, непристойные сцены или грубую ненормативную лексику.
   
    
* «**AO**» («Adults Only 18+») — «Только для взрослых»: Содержание игры только для взрослых старше 18 лет. Продукты из данной категории могут содержать длительные сцены непристойного содержимого.
   
      
* «**RP**» («Rating Pending») — «Рейтинг ожидается»: Продукт был отправлен в ESRB и ожидает присвоения рейтинга. Данный логотип используется только на рекламных презентациях и в демо-версиях игр до официальной даты выпуска в продажу.
   
     
* «**K-A**» - («Kids to Adults») — Содержание игры вполне подходит для возрастной категории от 6 лет и старше. Такие игры подходят для всех возрастов. Этот рейтинг был заменен на Everyone.

Заменим `K-A` на `E`, так как это устаревшая категория

```python
def replace(value):
    if value == 'К-А':
        return 'Е'
    else:
        return value

games_data['rating'] = games_data['rating'].apply(replace)
```

В этом столбце строкам с пропусками можно дать рейтинг no_ESRB, так как рейтинг ESRB используется для США и Канады, а выпускаются игры по всему миру.

```python
games_data['rating'] = games_data['rating'].fillna('no_ESRB')
```

#### Влияние отзывов от пользователей и критиков на продажи

Посмотрим, как отзывы влияют на продажи внутри одной платформы

Возьмем PlayStation 4

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

ps4_data = data_filtered[data_filtered['platform']=='PS4']

# корреляции
cor_with_critic_score = ps4_data['critic_score'].corr(ps4_data['total_sales'])
cor_with_user_score = ps4_data['user_score'].corr(ps4_data['total_sales'])

# график отзывов пользователей
ps4_data.plot(x='user_score', y='total_sales', kind='scatter', alpha=0.3, ax=ax1, grid=True)
ax1.set_title(f'Корреляция у пользователей: {cor_with_user_score:.2f}')
ax1.set_ylabel('Продажи (млн. проданных копий)')
ax1.set_xlabel('Отзывы пользователей')

# график отзывов критиков
ps4_data.plot(x='critic_score', y='total_sales', kind='scatter', alpha=0.3, ax=ax2, grid=True)
ax2.set_title(f'Корреляция у критиков: {cor_with_critic_score:.2f}')
ax2.set_ylabel('Продажи (млн. проданных копий)')
ax2.set_xlabel('Отзывы критиков')

plt.suptitle('Диаграмма рессеяния для PlayStation 4')
plt.show()
```

Видим, что коэффициент корреляции мнений критиков оказался положительным, а у пользователей — отрицательным. Мнение критиков обладает большим весом по сравнению с мнением обычных пользователей.

Соотнесём полученный результат с продажами игр на еще пяти актуальных платформах:

```python
fig, axs = plt.subplots(5, 2, figsize=(20, 25))
plt.subplots_adjust(wspace=0.7, hspace=3.5)
popular_platforms = filtered_sales.sort_values(('sum', 'total_sales'), ascending=False).index[1:6]

for idx, platform in enumerate(popular_platforms):
    platform_data = data_filtered[data_filtered['platform'] == platform]

    cor_with_critic_score = platform_data['critic_score'].corr(platform_data['total_sales'])
    cor_with_user_score = platform_data['user_score'].corr(platform_data['total_sales'])

    # график отзывов пользователей
    platform_data.plot(x='user_score', y='total_sales', kind='scatter', alpha=0.3, ax=axs[idx, 0], grid=True)
    axs[idx, 0].set_title(f'ПЛАТФОРМА: {platform}\n\nКорреляция оценок пользователей: {cor_with_user_score:.2f}', fontsize=18)
    axs[idx, 0].set_ylabel('Продажи (млн. проданных копий)', fontsize=14)
    axs[idx, 0].set_xlabel('Отзывы пользователей', fontsize=14)

    # график отзывов критиков
    platform_data.plot(x='critic_score', y='total_sales', kind='scatter', alpha=0.3, ax=axs[idx, 1], grid=True)
    axs[idx, 1].set_title(f'ПЛАТФОРМА: {platform}\n\nКорреляция оценок критиков: {cor_with_critic_score:.2f}', fontsize=18)
    axs[idx, 1].set_ylabel('Продажи (млн. проданных копий)', fontsize=14)
    axs[idx, 1].set_xlabel('Отзывы критиков', fontsize=14)

    
plt.tight_layout()
plt.show()
```

Значительное воздействие на объемы продаж наблюдается у платформ `WIIU` и `3DS` за счет отзывов пользователей, в то время как для `PlayStation 3` это влияние является наименее существенным. Стоит отметить, что мнение экспертов, хоть и не показывает превосходных результатов в общем, превышает важность пользовательских оценок. Исходя из данной взаимосвязи, **можно предположить, что это явление может оказать значительное влияние на продажи в будущем.**

### 5. Проверка гипотез

#### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Сформулируем нулевую и анльтернативную гипотезы:

- Н0: Средний пользовательский рейтинг платформ Xbox One и PC одинаковый
- H1: Средний пользовательский рейтинг платформ Xbox One и PC различается

```python
alpha = 0.05

x_one = data_filtered.loc[data_filtered['platform'] == 'XOne', 'user_score'].dropna()
pc = data_filtered.loc[data_filtered['platform'] == 'PC', 'user_score'].dropna()

results = st.ttest_ind(
    x_one, 
    pc,
    equal_var=False,
    alternative='two-sided')

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую минуту')
```

Нулевая гипотеза не отвергнута.

Проверим дополнительно нашу гипотезу, выведем средние пользовательские рейтинги и по платформам

```python
mean_rating_platform = data_filtered.groupby('platform')['user_score'].mean()

pc_mean_rating = mean_rating_platform.loc['PC']
print(f"Средний рейтинг для PC: {pc_mean_rating}")

xone_mean_rating = mean_rating_platform.loc['XOne']
print(f"Средний рейтинг для XOne: {xone_mean_rating}")
```

Видим, что рейтинги платформ Xbox One и PC статистически значимо не отличаются и находятся в районе 6.2-6.5

#### Средние пользовательские рейтинги жанров Action и Sports разные

Сформулируем нулевую и альтернативную гипотезы:

- H0: Средние пользовательские рейтинги жанров Action и Sports одинаковые
- H1: Средние пользовательские рейтинги жанров Action и Sports отличаются между собой

```python
alpha = 0.05

action = data_filtered.loc[data_filtered['genre'] == 'Action', 'user_score'].dropna()
sports = data_filtered.loc[data_filtered['genre'] == 'Sports', 'user_score'].dropna()

results = st.ttest_ind(
    action, 
    sports,
    equal_var=False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую минуту')
```

Также дополнительно проверим средними пользовательскими рейтингами для жанров

```python
mean_rating_genre = data_filtered.groupby('genre')['user_score'].mean()

sports_mean_rating = mean_rating_genre.loc['Sports']
print(f"Средний рейтинг для Sports: {sports_mean_rating}")

action_mean_rating = mean_rating_genre.loc['Action']
print(f"Средний рейтинг для Action: {action_mean_rating}")
```

Действительно, рейтинги разные: у Sports - 5.2, а у Action - 6.8

Отвергаем нулевую гипотезу. Обнаружены статистически значимые различия между средними пользовательскими рейтингами для жанров Action и Sports

#### Пояснение

В нашем случае:

Нулевая гипотеза (H0) утверждает, что средние пользовательские рейтинги жанров Action и Sports равны друг другу и что средние пользовательские рейтинги платформ Xbox One и PC равны друг другу.

Альтернативная гипотеза (H1) предполагает, что средние пользовательские рейтинги жанров Action и Sports отличаются между собой и что средние пользовательские рейтинги платформ Xbox One и PC отличаются между собой.

Формулируем нулевую и альтернативую гипотезу, чтобы понять, следует ли отклонить нулевую гипотезу на основе имеющихся данных.

Применение t-критерия (теста Стьюдента) для сравнения средних может быть обосновано в вашем случае по следующим причинам:

- Выборки, которые мы сравниваем (пользовательские рейтинги), являются независимыми.
- Распределение пользовательских рейтингов в выборках можно считать приблизительно нормальным.
- Дисперсии в выборках, возможно, схожи или учтены с корректировкой при использовании t-теста.

Таким образом, применение t-критерия в данном контексте обосновано для проверки гипотезы о равенстве средних рейтингов в двух независимых выборках. Тест Стьюдента позволяет оценить статистическую значимость различий средних значений и помочь в принятии решения относительно нулевой гипотезы.

**Топ-5 жанров**

Жанры в топах разных регионов повторяются, меняется лишь их распостраненность среди пользователей. Разница может быть обусловлена менталитетом и предпочтениями. Так три жанра входят в топ-5 всех трех регионов: Action, Shooter и Role-playing. Жанр Action, входящий в топ-5 всех исследуемых регионов, максимально популярен в Северной Америке. А в Японии любят Role-playing и Fighting. 

В тройке самых прибыльных жанров: Shooter, Sports, Platform, а самый неприбыльный — Adventure.

**Топ-5 рейтинга**

Топ-5 рейтинга: M (для взрослых), no_ESRB (игры без рейтинга, выпущенные не в Северо-Американском регионе), E (для всех), T (подросткам) и E10+ (от 10 лет и старше). Во всех регионах, кроме Японии, по объёмам продаж лидируют игры с рейтингом M. В Японии лидируют продажи игр с без рейтинга, так как многие игровые проекты не попадают под рейтинговую систему, которая используется для США и Канады.

Самые продоваемые игры имеют рейтинг M. С его маркировкой во всех анализируемых регионах продано 371.5 млн. копий игр. Игры без рейтинга (выпущенные не в Северо-Американском регионе) находимся на втором месте по продаваемости.
